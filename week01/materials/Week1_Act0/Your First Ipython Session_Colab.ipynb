### "Magic" commands
iPython includes "magic" commands, which start with a % sign. These are the important enhancements that IPython offers compared to the standard Python shell.

#### %history

You can get the IPython history, which Jupyter Notebook uses as its kernel, by running the `%history` command. This command dump a history of every command you have run in the current IPython session. Sometimes it is probably more than you want, but it is better than losing all your work.

````python
%history
````

You can also add the corresponding line numbers by using the optional flag `-n`.
Execute the following command and compare the difference with the previous output.

````python
%history -n
````

#### %rerun
Once you've printed out the IPython history with the indicated line numbers, you can easily rerun a command using the `%rerun` command. 
For instance, the following will cause the item number 2 from your session history to be executed, which, would say hello to the world again!
You can also scroll through your input history using the up and down arrow keys.

````python
%rerun 2
````

### Interactive Help in Python
Interactive help is available in IPython through the `?` and `??` operators. This can display information on functions and defined objects. For example, `?range` will display information about the `range` type in Python (The range type represents an immutable sequence of numbers and is commonly used for looping a specific number of times in for loops. see [Python doc: Ranges](https://docs.python.org/3/library/stdtypes.html#typesseq-range)). 

````python
?range
````

Also, if you have defined a variable previously in this session, you can use `??` to show the information about the variable. 
Using `??` instead of `?` can display extra information for some objects.
For example, we have defined the `name` and `job` variable before, so we can retrieve the information of them below.

````python
??name
````

````python
??job
````
