- Load with the `import` statement: `import <module_name> ` or `import <module_name> as <alias>` keywords, e.g.:

````python
import pandas as pd 
# indicate that we are importing pandas library
# pd is the name we are giving to the imported library. 
````

- Sometimes rather than importing the whole module, you may want to import just a specific function that you want to use. You might encounter some memory constraint that do not allow you to import the whole module. In such cases, the `import` statement will look like the following: 
    - `from <module_name> import <function_name> as .<alias>`
    - `import <module_name>.<function_name> as <alias>`

For instance:

````python
from matplotlib import pyplot as plt
# Can also use shorter “.” notaion rather than “from” to load library: import matplotlib.pyplot as plt
# plt is the name we are giving to the imported library. 
# We use it whenever we want to call a function provided by the library, e.g., plt.boxplot(…) 
````

## 1. Data Tables in Python
The Pandas library provides a table structure called DataFrame. A DataFrame is a two-dimensional, size-mutable, potentially heterogeneous tabular data. More information about Pandas DataFrame can be found [here](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.html).

### 1.1 Creating DataFrame

You can create a Pandas DataFrame as follows.

````python
# create the dataframe
df = pd.DataFrame({
 'StudentID' : [264422,264423,264444,264445,264446], # StudentID column and its data
 'Name' : ['Steven','Alex', 'Bill', 'Mark','Bob'], # Name column and its data
  'EnrolYear' : [2010,2010, 2011, 2011,2013], # Enrolment year column and its data
  'Math' : [100,90,90,40,60], # Mathematics score column and its data
  'English' : [60,70,80,80,60] # English score column and its data
})

# print out the dataframe
df
````

### 1.2 Basic Element Selection

You can perform some element selections with Pandas DataFrame, such as selecting a column, selecting a row meeting some conditions, selecting cells with complex conditions, generating new columns.

````python
# Select a column by using its column name:
df['Name']
````

````python
# Select row when it meets conditions 
df[df['Name'] == 'Alex'] # select record of Alex
````

````python
# Selection with complex conditions
filt = (df.Name != 'Bob') & (df.StudentID > 264423) # select rows where name is not 'Bob' and student ID is larger than 2644423
df[filt]
````

````python
# Operation on columns
# Create a total score column
df['Total'] = df['Math'] + df['English']
df
````

### 1.3 Melting Data Table
In the table above, we can see that each subject (E.g. *English*, *Math*) is in different columns, so we can melt the table to make these subjects in one column by using the `melt()` function. We can also rename the column via the `rename()` method.

````python
#columns named 'Math' and 'English' are now values of column 'Subject'
df = pd.melt(df, id_vars=['EnrolYear','Name','StudentID'],value_vars=['Math','English'],var_name='Subject')

# you can also rename a column seperately:
df.rename(columns = {'value':'Score'}, inplace = True)
df
````

### 1.4 Merging Data Table
Now assume if you got another data table storing the heights of the students and suppose now for some reasons you want to merge two data tables. You can then use the `merge()` function as follows.

````python
# notice in df2, 'Bob' (ID: 264446) is missing
df2 = pd.DataFrame({
 'StudentID' : [264422,264423,264444,264445], # StudentID column and its data
  'Height' : [160,155,175,175], # Height column and its data
})

df2
````

````python
merge = pd.merge(df,df2, on=['StudentID'])
merge # have a look
````

## 3. Group by
Continue with the Titanic data set, but now let's focus on the ages of passengers. 
We showed how to look at the mean of column using the `describe()` function (alternatively, you can get the mean by `titanic['age'].mean()`).
Here we will go for a more detailed look at the ages of different groups.

````python
# this groupby has created a series with a hierachical index. We can then apply methods to the individual groups
sex_class = titanic.groupby(['sex','class'])['age']

# have a look
sex_class.mean()
````

In a more sophisticated case, if you want to do statistics on multiple columns or if you want to apply different aggregate functions to different columns, this is one way to go....

````python
# In the following, we specify aggregation functions to columns 'who' and 'age', grouping by the 'class' column 
# we've also created the columns 'passengers' and 'average_age' to store the statistics values
# the number of records in 'who' is named as 'passengers', and,
# the mean values of 'age' is called 'average_age'
groupbyClass = titanic.groupby('class').agg(passengers = ('who','count'),
                                            average_age = ('age','mean'))

# The alternative way to do this is as follows:
# titanic.groupby('class').agg({'who': 'count', 'age': 'mean'}).rename(columns={'who': 'passengers', 'age': 'average_age'})


groupbyClass
````

````python
# we try to make it look pretty, i.e., make it as a dataframe, you can have a look at it by yourself step by step
groupbyClass = groupbyClass.reset_index() # reset its index


# now have a look, so this tells statistic about the average age and the number of passengers in each class
groupbyClass 
````

## 4. Applying Function
Applying function is a built-in function in Python, it is a very handy tool when we want to apply a function to each element of a list, series, or rows/columns of Dataframes etc.

It is also very handy when we combine the use of the `lambda` function (also called anonymous function: a function that is defined without a name).

````python
# here, we want to see: the age diversity in each class; and the number of people elder than 50 in each class
# so in the following, the two lamda functions apply to the age column, 
# the 'unique age values' counts it unique not null values, 
# and the 'elder age group' counts the number of values larger than 50
# notice that here we apply two functions to the same column 'age'.

groupbyClass = titanic.groupby('class').agg(unique_age_values = ('age',lambda x: x.nunique()),
                                            elder_age_group = ('age',lambda x: x[x>50].count()))

# the alternative way to do this is as follows
# groupbyClass = titanic.groupby('class').agg(Lambda1=('age', lambda x: x.nunique()), Lambda2=('age', lambda x: x[x>50].count())).rename(columns={'Lambda1': 'unique_age_values', 'Lambda2': 'elder_age_group'})
# groupbyClass

groupbyClass = groupbyClass.reset_index() # reset its index
groupbyClass # now have a look
````

## 5. Basic Plotting

### 5.1 Simple Plot 
You can just simply plot the values to have a look. But sometimes this wouldn't tell you any meaningful insight. For example:

````python
import matplotlib.pyplot as plt #import library for plotting
# This following magic line is to make sure graph shown inline of the notebook
%matplotlib inline
# continue to use titanic data set
plt.plot(titanic.fare) # this simply plots the fare of each passenger, so there are about 900 passengers.
plt.show()
````

### 5.2 Histogram
Instead, we could have a look at the distribution of the data by plotting the histogram. 

````python
titanic.fare.hist(bins = 200) # you can try different bin number to have a look
plt.xlim(0, 300) # setting limit on x-axis
plt.ylim(0,600) # setting limit on y-axis
````

### 5.3 Boxplot
Or alternatively, we can use boxplot, which helps you to identify any outliers. A boxplot (also called a box and whisker diagram) is a simple visual representation of key features of a univariate sample. 
It displays five-point summaries and potential outliers in graphical form. 

The following figure shows an illustration graph of boxplot elements.

<img src="Boxplot.png"></img>

````python
titanic.boxplot(column = 'fare')
plt.ylim(0, 600) # setting limit on y-axis
# we can see that there are lots of outliers in column 'fare'
````

We can see that there are lots of outliers in the column `fare`, but it probably makes much more sense to look at it by different classes:

````python
titanic.boxplot(column = 'fare', by = 'class')
plt.ylim(0, 600) # setting limit on y-axis
# we can see that there are lots of outliers in column 'fare'
````

Now, we can remove the outliers in different classes. Notice that you could have made the following code more efficient and accurate. But we are trying to keep the simplicity here.

````python
# deleting outliers for First class
filt = ~((titanic['class'] == 'First') & (titanic['fare'] > 160))
titanic = titanic[filt]

# deleting outliers for Second class
filt = ~((titanic['class'] == 'Second') & (titanic['fare'] > 50))
titanic = titanic[filt]

# deleting outliers for Third class
filt = ~((titanic['class'] == 'Third') & (titanic['fare'] > 30))
titanic = titanic[filt]

# Now have a look at it, should be much better!
titanic.boxplot(column = 'fare', by = 'class')
plt.ylim(0, 600) # setting limit on y-axis
# we can see that there are lots of outliers in column 'fare'
````

### 5.4 Barchart
We can compare the fares for different classes and for kids/adult by bar chart. 
To do that we need to look into the fare for adule (age >= 18) and child (age < 18) seperately. 
In the following, we use pandas `groupby` based on `class`, and we do an apply function on the rows of in each group, so that we can split into two different categories, `child_fare` and `adult_fare`:

````python
# prepare the statistics of children fare (under 18) and adult fare in different classes

groupbyClass = titanic.groupby('class').apply(lambda x: pd.Series(dict(child_fare = x[x['age']<18]['fare'].mean(),
                                                                       adult_fare = x[x['age']>=18]['fare'].mean())))

groupbyClass = groupbyClass.reset_index() # reset its index
# now have a look, so this tells statistic about the average age and the number of passengers in each class
groupbyClass 
````

Now, we can compare the difference between adult fare and child fare using bar chart:

````python
import numpy as np
ind = np.arange(0, 3*2, 2)
width = 0.5      # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(ind, groupbyClass['child_fare'], width, color='g')
rects2 = ax.bar(ind + width, groupbyClass['adult_fare'], width, color='c')

# add some text for labels, title and axes ticks
ax.set_ylabel('fare')
ax.set_title('Child and Adult fare for different classes')
ax.set_xticks(ind + width+0.1)
ax.set_xticklabels(groupbyClass['class'],rotation='vertical')

ax.legend((rects1[0], rects2[0]), ('child_fare', 'adult_fare'))

fig.set_size_inches(18.5, 10.5)
````

Oops, that looks a bit suspcious, how come child fare is more expensive than adult?
Let take a the 'First' class as example, and look at some summary statistics there:

````python
# summary statistics of 'First' class fare for children under 18
titanic[(titanic.age<18) & (titanic['class'] == 'First')].fare.describe()
````

Indeed, 75% of child fare is more than 69.9, and the median is 108.9. 
Let's have a look at the adult fare:

````python
# summary statistics of 'First' class fare for adult above 18
titanic[(titanic['age']>=18) & (titanic['class'] == 'First')].fare.describe()
````

Ah ok, that's consistent. The adult fare seems to be cheaper. 
It's hard to guess a reason though, but that's what the data tells us. 
Indeed, if you try the price for younger age children, e.g., under 5, you might realise the fare for them is expenseive.

````python
titanic[(titanic['age'] <= 5 ) & (titanic['class'] == 'First')]
````
