## 1. Overview of Python

### 1.1 General Purpose Language
Python is a general-purpose programming language that can be used to literally build programs in a wide variety of application domains.
It can be used for developing relatively sophisticated programs, ranging from data analytics and website development to gaming and robotic applications.

### 1.2 History?
Python is not a new programming language and it has been around for a long period of time (over two decades!?!). 
It was first conceived in the late 1980s by a Dutch programmer, [Guido van Rossum](https://en.wikipedia.org/wiki/Guido_van_Rossum). 
The language was named after Guido's personal favourite comedy series "Monty Python's Flying Circus".

### 1.3 The Love of Data Scientists
Python is the fastest-growing programming language in terms of active developers, according to the annual [Stack Overflow Developer Survey](https://insights.stackoverflow.com/survey/2019). 
It is notably one of the most popular languages used in data science.
Many of us (especially those who enjoy playing with data) would have loved Python as it is considerably a simple programming language that is easy to learn and use. 
It also comes with the support of a rich collection of libraries (or packages). 


### 1.4 Fast Evolving
Python is in fact a living language that constantly evolves since its first introduction. There are two major versions of Python (https://www.python.org): Python 2 and Python 3.
*   Python 2.0 was released on 16 Oct 2000 with many subsequent 2.x versions. 
Then in Dec 2008, Python 3.0 was released to mainly address various design decisions and inconsistency in Python 2 and its subsequent releases.
*   Python 3 is strongly recommended for any new development. 
According to Python.org, as of January 2020, Python 2 has reached End Of Life status, meaning it will receive no further updates or bugfixes, including for security issues.

Due to the fast evolving nature, you may have seen and will see many subsequent releases 3.x (and for each many 3.xx as well). 
On the one hand, this is a good thing since the community are actively using, developing and improving it. 
On  the other hand, it presents its own challenges, e.g., it may present some uncompatibilities in what we have coded. 
Though, this is unavoidable and we all need to keep up with it! Be ready! :)

### 2.1 Some Basic Notebook Shortcut Keys
There are two 'modes' in the Jupyter Notebook. 

- Hit the *'esc'* key on your keyboard or click anywhere outside the cell to enter the **command mode**. The command mode can be identified by the grey border around the cell and the blue left margin.
- Hit the *'return'*/*'enter'* key on your keyboard or click in a cell to enter the **edit mode**. In this mode, the cell border turns green.

The command mode lets us talk to the Notebook whilst the edit mode is where we write code in the cell.

Finally,
- Hit *'shift'* + *'return'*/*'enter'* or press the *'Run'* button in the toolbar above to run the code. They are side by side on the right of the keyboard. 

### 2.2 In the Command Mode (Grey Border with Blue Left Margin)

## 3. Markdown
Markdown is how you can write text, create links, show images and videos.
Markdown is ironically a mark-up language. You might know other mark-up languages like HTML, where a small number of character combinations are used to indicate the formatting of the text.

Markdown's set of formatting characters are quite unobtrusive and take little effort to write and don't get in the way of reading it, even in it's raw (unformatted) state.

Learning how to use Markdown in Jupyter Notebook is important, and it could make it handy for you to structure and format your notebook nicely, increasing the readability of your work. 

### 3.1 Headings
The *'#'* key will create headings, with *'#'* being the biggest up to *'######'* being the smallest. In the following markdown cell, we are playing with different level headings. See how they are written, and how they would show up. 

=====================================
# 1st level heading
## 2nd level heading
### 3rd level heading
#### 4th level heading
##### 5th level heading
======================================

### 3.2 Bullet Points 
The *'-'* will create a bullet point. You can try to run the following Markdown cell to see the effect.

- Bread

- Cheese

### 3.3 Numbered Lists
Replace the *'-'* character with numbers (the numbers don't have to be in order), to get a numbered list. You can try to run the following Markdown cell to see the effect.

### 4.1 Readability counts!!!

As a data scientist you are communicating your insights to others. The reader may not be primarily a programmer. They may work in another language and they may not be a machine learning practitioner. This is why readability is important.

We use Python and Jupyter Notebook because they enhance **readability** and your ability to clearly explain your insights. It is important to **put yourself** in the position of **your readers**. You should explain your process so they don't have to spend too long working out what you have done and why.

Interestingly, there are a few good programming concepts that conflict with this readability.

**Declaration of variables and functions**
One practice is to define lots of variables and functions at the start of the script. While this makes sense from a code development point of view because there is one place to change a value that might be used throughout the script. But from the point of view of the readers, it means they have to remember what the value of that variable is until it is used, maybe dozens of lines further down in the script. A long with defining the variables in the same section as the code, meaningful variable names will help the readers follow your process. Also, do not save many useless dataframes, which confuse your readers and make your code slow too.

**Try to be simple, but efficient**
Another area to consider carefully is not try to amaze your readers with complicated solutions. Think twice when you are writing functions or even classes to perform actions. Simplicity is better than complexity, and complexity is better than complication. Sometimes complex solutions can't be avoided because the problem is complex but you need to consider if a step-by-step procedural approach might be clearer for the readers than an object-oriented approach. Again, if you do define functions, defining them just before you use them will be easier on your readers than defining all functions at the start or the end of the script and making them remember how the function operates. 

**Using existing libraries or functions**
On that same topic. Try to use the built-in functions of the major libraries rather than writing your own. It is difficult for an individual to write code that is as optimised and versatile as the group that has developed the library. Many open source projects have hundreds of contributors and reviewers. Remember, *'Many eyes make all bugs shallow'*. Take advantage of the work the community has done writing documentation and having a consistent interface. Scikit-learn is a great example of this. There is also the possibility that your readers might be familiar with the built-in functions of the library you are using and will therefore understand what you are doing with less effort than if you create your own way to do the same task.

One more thing: **"beautiful is better than ugly"**. Use headings and sub-headings in the Markdown cells and an introduction to guide your readers. Do not put all your code in one cell. A cell should be a self-contained function unit. 

Some additional informaiton: it is a good practice to structure your notebook, such as: 

* Introduction
* Notes on Data source, output data, versions of your program etc. 
* Data preparation steps
* Analysis (possibly with visualisation of the results)
* Discussion 
* Conclusions
